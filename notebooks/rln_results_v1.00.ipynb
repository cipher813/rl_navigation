# RL_Navigation Results

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATH = "/Volumes/BC_Clutch/Dropbox/DeepRLND/rl_navigation/"
```

```python
def chart_results(PATH, pklfile):
    CHART_PATH = PATH + "charts/"
    pklpath = CHART_PATH + pklfile
    timestamp = pklpath.split(".")[-2].split("-")[-1]
    chartpath = CHART_PATH + f"NavigationTrainChart-{timestamp}.png"

    with open(pklpath, 'rb') as handle:
        scores = pickle.load(handle)    
    # plot the scores
    print(f"Environment solved in {len(scores):d} episodes.\tAverage Score: {np.average(scores[-100:]):.2f}")
    fig = plt.figure()
    ax = fig.add_subplot(111)
    plt.plot(np.arange(len(scores)), scores)
    plt.ylabel('Score')
    plt.xlabel('Episode #')

    plt.savefig(chartpath)
    print(f"Chart saved at {chartpath}")
    plt.show()
    return scores
```

```python
scores = chart_results(PATH, "NavigationTrainChart-201811231149.pkl")
```
